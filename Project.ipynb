```python
rd = pd.read_csv('orders.csv',na_values=['Not Available', 'unknown'])
rd['Ship Mode'].unique()
rd.head(10)
```

```python
rd.rename(columns={'Order Id': 'Order_Id'})
```

```python
#rd.columns=rd.columns.str.lower()
rd.columns=rd.columns.str.replace(' ', '_')
rd.columns
rd.head(10)
```

```python
rd['discount'] = rd['List_Price']*rd['Discount_Percent'] / 100
rd['sale_price'] = rd['List_Price']- rd['discount']
rd['profit'] = rd['sale_price']- rd['cost_price']
rd
```

```python
rd.drop(columns=['cost_price','List_Price','Discount_Percent'],inplace = True)
rd
```
